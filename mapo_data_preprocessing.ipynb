```python
# 관광협회 지정
tour = pd.read_json(json_files[1])
tour['지정기관'] = '서울시관광협회지정 관광식당'
tour[:2]
```

## 마지막: 주소에서 중복 제거
- string 반으로 잘라서 same check
- list로 split하고 in 체크로 중복 제거후 join

```python
print(total['위치'])
```

```python
processed_adrs = []
for v in total['위치']:
    adrs = v.split(' ')
    post = []
    for a in adrs:
        if a not in post:
            post.append(a)
    print(' '.join(post))
    processed_adrs.append(' '.join(post))
```

```python
total['위치'] = processed_adrs
```
